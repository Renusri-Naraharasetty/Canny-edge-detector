```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread("boy.jpg",cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img,cmap='gray')
```

```python
blr=cv2.blur(img,(3,3))
plt.imshow(blr, cmap="gray")
```

```python
medium=np.median(blr)
medium
```

```python
low=int(max(0,0.7*medium))
high=int(max(255,1/3*medium))
```

```python
low
```

```python
high
```

```python
canny_blr=cv2.Canny(blr,low,high)
```

```python
plt.imshow(canny_blr)
```

```python
edge=cv2.Canny(img,115,115)
plt.imshow(edge)
```
